```python
#now we are gonna make the columns to come into right format: -
#firstly we start with genres: -
movies.iloc[0].genres#The movies.iloc[0].genres expression is used in Python with pandas DataFrames to access the value(s) in the "genres" column of the DataFrame for the first row (index 0).
```

```python
#'[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}] 
#hum chahte hai ki oper vala column simplify hokee kuch aessa dikhe: -{action,adventure,fantasy,science fiction} yeh ek tarah se sirf genres dekhaega movie ke.
#for this we gonna take the help of the helper function.
import ast
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):#with the help of this method humme hmari list mill jaegi from string form.
        #ast.literal_eval(obj) is a method in Python provided by the ast (Abstract Syntax Tree) module. It is used to safely evaluate a string containing a single Python literal (i.e., a value that can be expressed directly in Python code) and return the corresponding Python object.
        L.append(i['name'])#basically humme sirf genres nikalne hai os puri dectionary mai se
    return L
movies['genres']=movies['genres'].apply(convert)
movies.head()
```

```python
#aesse he simplify karenge hum ab: - keywords column ko
movies['keywords']=movies['keywords'].apply(convert)
movies.head()
```

```python
#hum same oper valle chizz karenge cast ko simplify karne ke liye or es ke ander bhi hum sirf phele 3 or main actors he dekhaenge hur movie ke liye.
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):#with the help of this method humme hmari list mill jaegi from string form.
        #ast.literal_eval(obj) is a method in Python provided by the ast (Abstract Syntax Tree) module. It is used to safely evaluate a string containing a single Python literal (i.e., a value that can be expressed directly in Python code) and return the corresponding Python object.
        if counter!=3:
            L.append(i['name'])#basically humme sirf genres nikalne hai os puri dectionary mai se
            counter+=1
        else:
            break
    return L
#above whole function is just for extracting the first three main heroes from the cast column
```

```python
movies['cast']=movies['cast'].apply(convert3)
movies.head()
```

```python
movies.info()
```

```python
#now we gonna simplify the last column which isd crew and in that also we want to extract the director name column only becasue we wqant only director to be reflected in that crew whole column.
def fetch_director(obj):
    L=[]
    for i in ast.literal_eval(obj):#with the help of this method humme hmari list mill jaegi from string form.
        if i['job']=='Director': 
        #ast.literal_eval(obj) is a method in Python provided by the ast (Abstract Syntax Tree) module. It is used to safely evaluate a string containing a single Python literal (i.e., a value that can be expressed directly in Python code) and return the corresponding Python object.
            L.append(i['name'])#basically humme sirf genres nikalne hai os puri dectionary mai se
            break
    return L
```

```python
movies['crew']=movies['crew'].apply(fetch_director)
movies.head()
```

```python
#as our overview column is in string form let's also convert it in list form like other: -
#hukm ess overview valle block ko bhi list main eslie convert kar rhe hai takii osse bhi bakii list containing columns ke sath concatenate kar pae.
movies['overview']=movies['overview'].apply(lambda x:x.split())
```

```python
#now we arer gonna merge all the columns from overview to crew and oske baad onko list se string main le aaenge feer hmaree pass ek bda sa paragraph ban jaega .and that whole paragraph is our whole tag column.
#there is one transformation which we have to apply which is based on removing all the spaces in between the words because: -
#yeh eslie kar rhe hai kyu ki jab hum onhe list se string main convert karenge then manna ek naam that par voh alg alg hokjee two different entities ban jaenge example; - [sam mendes]=sam alg and mendes alg
#oper valle chizz se problem aaenge jab hmare recommendation sysytem main log search karenge sam mendes toh onke pass alg alg sam aaenge or kya pta kisi or sam ki movie aajae jab ki onhe dekhni koi or thae toh os mistake ko htane ke liye hum combine kardenge names like sammendes
#jisse onhe onki recommendation valle movies mera model de paega:
#our transformation is: -
#for genres: -
movies['genres']=movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])#esse basically humne space valli jgah pe(" ") no space dalni hai("")
movies['keywords']=movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast']=movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew']=movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
#ab hum ek column bna rhe hai [tags] naam ka jiske ander hum apne overview se lekar end tak ke saree columns ko add karlenge
movies['tags']=movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
movies.head()
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():#from here we are converting the string into the list.
        y.append(ps.stem(i))#with this we are going to stem every word.
    return" ".join(y)#now we are converting the list into the string again.
```

```python
new_df['tags']=new_df['tags'].apply(stem)#esse ab hmare sarree niche valle output ophele se differnt aaenge basically all the words according to their root node word .because we apply stemming here.
```

```python
#as wee see based on the basis of text of the two differtent movies we have to find the similarity in between them :-this is our chaLLENGE
#more the similarity in between the movie means we have to find more similarity in the text also.

#hum basically similarity eslie dhondh rhe hain takii jiss type ki movie pasand kar rha hai user usse type ki movie usse mille

#just because of this problem the concept of vectorisation comes on which we convert the text into vector form.

#ek tarah se hum hur movie ko ek vector assign kar rhe hai jisme agr humme ek movie dekh rhe hai toh recommendation sysytem ko or konsi movies mujhe recommend karni chahiye this depends on the genre and just according to that it will recommend me the closest vector of the movie which i am currently watching.

#we are using the [BAG OF WORDS] TECHNIQUE OF TEXT VERCTORISATION CONCEPT: -
#IN THIS TECHNIQUE WE ARE BASICALLY COMBINING ALL THE TAGS AND THROUGH THAT COMBINED TAG WE HAVE TO EXTRACT THAT MOST COMMON REPITITIVE WORDS FROM THAT COMBINED TAG.
#AFTER DOING THIS I CHECK THAT HOW MANY TIMES THIS MOST COMMON REPITITIVE WORDS COMES IN THE INDIVIDUAL TAG OF THE MOVIE.WE DO THIS FOR EVERY MOVIE.
#AND AT THE END OF THIS WE GET A TABLE WHOSE SHAPE IS((MOVIES=5000,(MOST COMMON WORDS=5000))

# BASICALLY IN THIS WHOLE IMFORMATION AT THE END OUR VECTORS ARE THOSE WHICH ARE BUILD BY COUNTING THE MOST COMMON WORDS IN INDIVIDUAL MOVIE TAGS AND OSSE JOH MOST COMMON WORDS KE COUNT SE JOH VECTOR TAYAR HOGAA THAT IS THE ONE VECTOR.
#NOTE:- IN THIS VECTORISATION CONCEPT WE HAVE NOT TO COUNT THE STOP WORDS WE HAVE TO GIVE ASIDE TO THEM(NOT INCLUDE THEM)
#STOP WORDS ARE THOSE WHICH ARE USED IN SENTENCE FORMATION BUT THEY ARE NOT USED IN OR THERE IS NO CONTRIBUTION OF THEM IN SENTENCE BUILDING.EG: - IS,ARE,AND,OR,TO,FROM AND OTHERS.
```

```python
#as we are talking about 5000 dimensional space in which every movie is a vector.
#no we are gonna calculating the distance of every movie vector with respect to another movie's vector. #more the distance lesser will be the similarity and less the distance more will be the similarity.
#here we are not gonna calculate the euclidian distance infact we are here using and calculating the cosine distance.

#basicaly in cosine distance we are gonna calculating basically the angle between the two movies lesser then angle mmore will be the similarity.

#we are using the cosine distance because in high order dimensionality like 5000 which we are using write now eucliedian distance doesn't work very well due to the reason of (curse of dimensionality).

# The following are some of the specific implications of the curse of dimensionality for Euclidean distance:

# The average distance between two points in high-dimensional space increases rapidly with the number of dimensions. This means that most points are likely to be very close to each other, regardless of their actual similarity.
# The distribution of distances between points in high-dimensional space becomes more and more uniform. This means that it becomes difficult to distinguish between similar and dissimilar points based on their Euclidean distance.
# Euclidean distance becomes less and less informative about the similarity of points in high-dimensional space. This is because there are many different ways to travel between two points in high-dimensional space, and the shortest path may not be the most meaningful.
# As a result of these factors, Euclidean distance is often not a good choice for measuring similarity in high-dimensional spaces. There are a number of other distance metrics that are better suited for high-dimensional data, such as cosine similarity and Jaccard similarity.

# for finding the cosine similarity(as distance is inversely proportuional to similarity) we are using sckit learn library for importing one function name cosine similarity.
from sklearn.metrics.pairwise import cosine_similarity#cosine similarity has two values 0 and 1 noly.
cosine_similarity(vectors)
```

```python
similarity=cosine_similarity(vectors)
similarity#if we see carefully that in output we are getting array of arrays means hur ek movie ke ander humme oss movie ka bakii movies ke sath ka similarity mill rha hai in array form.
```

```python
#now we are making a function jiske ander hum kuch aessa karenge ki agr humne ek movie daalli toh voh oske jaise he 5 movies mujhe recommend karke dega: -
#if we are given with any movie than first we have to find the index of that movie in the data.
#then we have to go in similarity matrix for that specific index movie 
#feer oss movie se saree movies ka distance nikalenge(eg: - avatar)
#now we sort the distance and we get the less distance movie at the first means more similarity and then we fetch that msot similar movies from the dataset.
def recommended(movie):
    #first we are going for the index:
    movie_index=new_df[new_df['title']==movie].index[0]
    distances=similarity[movie_index]#these distances are in array so we have to sort the array for getting the closest first five movies.
    #humne niche valli line kyu likhe: - 
    #phele toh yeh line humme phele 5 movies dhondhne mai help karegi.or hum sort ess taarh se karenge takii hmare orginal movie ka distance osse hmesha 1 rhega voh na change hoon kyu ki agr voh change hogya during sorting then indexing kharab hoon sakti hai
    #humme during sorting ek movie ka oska apne aap se indexing pakad ke rakhna pdega takii indexing lose na hoon or oske baad valli 5 movies par puraa kaaam karna hai. for this only we use the (enumerate function)
    #list main  karenge usse or then sorting and in reverse order takii smallest distance valli sabse phele aae.
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6] #(x:x[1]) it is used for telling that humme first valle number ke base pe nhi second valle ke base pe se sorting chalo karna hai.takii indexing maintain rhe
    
    for i in movie_list:
        print(new_df.iloc[i[0]].title)#with the help of this we get that specific index movies.
```

```python
recommended('Avatar')#witht this we get the index of the most similar movies according to the movie which we mentioned first.
```
